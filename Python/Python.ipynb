# 1. Number Guessing Game

1. Enter the range of numbers you want the target number to be between, and press New Target
2. Enter your guess
3. See if your guess is too high or too low
4. Change your guess
5. The app will tell you when you are right, and how many guesses you had

## 1.1 Method 1: Unlimited Retries

```python
secret_num = 54
while True:
    get_num = int(input("Guess the number: "))
    if get_num == secret_num:
        print("You guessed it right!")
        break
    else:
        print("Try again!")
```

## 1.2 Method 2: Three Retries

```python
count = 3
secret_num = 54
win = 0
for i in range(0,count):
    get_num = int(input("Hi,Guess the number: ")) if i==0 else int(input("Try Again Now.. Guess the number: "))
    if get_num == secret_num:
        win = 1
        print("You guessed it right")
        break
print("Oops, You've lost all the 3 attempts, Please try again tomorrow, Bye..!!") if win==0 else print("Well done, Good bye..!!")
    
```

## 1.3 Use Only The Given Range

#### Get and Validate Input Functions

```python
def get_input(text):
    while True:
        try:
            num = int(input(text))
            return num
        except ValueError:
            print("Oh no!, Entered input is invalid, Please try again with a Number")
```

```python
def validate_input(secret,guess):
    if guess==secret:
        return True
    elif guess<secret:
        print("Your Number is lesser than Secret Number, Please Try Again Now..")
    else:
        print("Your Number is greater than Secret Number, Please Try Again Now..")
    return False
```

#### Main Function

```python
import random
print("Please tell me the Range of Numbers you want to play..")
A = get_input("Enter Starting Number: ")
B = get_input("Till Number: ")
secret_num = random.randint(A,B)
total_attempts = 0
while True:
    try:
        get_num = int(input("Guess the Number: "))
        if A<=get_num<=B:
            if validate_input(secret_num,get_num):
                print(f"Yayy! You Guessed it Right, You've used {total_attempts+1} attempts to find the Secret Number. Well done, Good Bye..!!")
                break 
            total_attempts += 1
        else:
            total_attempts += 1
            print(f"Oh no!, That's not within the range, Please guess again a number between {A} and {B}")
    except ValueError:
        print("Oh no!, Entered input is invalid, Please try again with a Number")
            
    
```

# 3. Functions & Loops

## 3.1 Basics

```python
# Built In functions -> len(), abs(), random.randint()

# User defined Functions

def myfunc(name:str, a, b, c=2):
    print(f"Hi {name}!")
    return a+b+c
print(f'Your Sum is: {myfunc(name="Nithisha", b=4, a=5, c=100)}')


def myfunc(name:str, *args,**kwargs):
    print(f"Hi {name}!")
    ans = 0
    for arg in args:
        ans += arg
    ans_2 = ""
    for kw in kwargs:
        ans_2 += kwargs[kw]
    return ans,ans_2
print(f'Your Sum is: {myfunc("Hello", 1,1,2,3,2, first="Nithisha",last="Karthikeyan")}')


def factors(num):
    i = 1
    ans = []
    while i<=num:
        if num%i == 0:
            ans.append(str(i))
        i += 1
    return ans
a = 20
print(f'Factors of {a} are: {",".join(factors(a))}')
```
