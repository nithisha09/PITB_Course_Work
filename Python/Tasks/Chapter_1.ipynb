## Frequently used String Methods

```python
# Lower, Upper, Capitalize, Title
print(f'Lowercase: {"TamilNadu".lower()}; {"INDIA".lower()}')
print(f'Uppercase: {"india".upper()}; {"AsiA".upper()}')
print(f'Captialize: {"12india".capitalize()}; {"asia".capitalize()}; {"ASIA".capitalize()}')
print(f'Title: {"Tamil nadu".title()}; {"the atomic habits".title()}; {"234 constituencies".title()}')

fruit = "  blueberry "
veggie = "carrot   "

# Strip
print(f"Start&End Spaces: {fruit.strip()}; {veggie.strip()}")
print(f"Leading Spaces: {fruit.lstrip()}; {veggie.lstrip()}")
print(f"Trailing Spaces: {fruit.rstrip()}; {veggie.rstrip()}")

# Replace
print(f'Replace: {"blueberry".replace("berry","star")}, {"watermelon".replace("123","fall")}')

# String <-> List
print(f"String to List: {'one:two,three'.split(':')}")
print(f"String to List with Limits: {'one:two,three'.split(':')}")
print(f"List to String: {'-'.join(['i','c','e'])}")

# Find, Count
a = "India is in Asia"
print(a.find("Africa"))
print(a.count("India"), a.count("Africa"))

# Startswith, Endswith
print(a.startswith("Ind"), a.endswith(" "))

# isalpha, isdigit
name = "123"
print(name.isalpha(), name.isdigit())
```

## Frequently used String Functions

```python
# len
print(len("Iodine"), len(""))

# str()
print(str(10), type(str(10)))

# Char to Ascii
print(ord("A"), ord("a"), ord("0"), ord("9"))

# Ascii to Char
print(chr(65))

# Sorting to List
print(sorted("kij"))

# min, max
print(min("0a"), max("0a"))
print(any(""), any("i"), any([1,0]), any([0,0]))
print(all([1,2]), all([1,0]))

#enumerate
s = "sock"
for i,v in enumerate(s):
    print(f"Index:{i}, Value:{v}")

# str to list
print(list(s))

# str to tuple
print(tuple(s))

# ascii -> non-ascii to unicode
print(ascii('Pythön')) 

# reverse
rev = reversed("bin")
# for i in rev:
#     print(i)
print("-".join(reversed("bin")))
```

```python
# while loop, break
print("Never repeat the same letter")
inputs = ""
get_input = ""
while True:
    get_input = input("Enter a letter: ")
    if inputs.count(get_input)!=0:
        break
    inputs += get_input
    print(inputs)
print(f"Oops! You did enter letter {get_input} again")

# for loop
# if else
# if elif else
marks = [90, 95, 80, 75, 35]
grade = None
for i in range(0,len(marks)):
    if marks[i]>=90:
        grade = "A"
    elif marks[i]>=80:
        grade = "B"
    elif marks[i]>=70:
        grade = "C"
    else:
        grade = "D"
    print(f"Index: {i}, Marks: {marks[i]}, Grade: {grade}")

    
def greet_user(name):
    print(f"Hello {name}, Welcome to Chennai")
greet_user("Mate")

def greet_team(team):
    if len(team)<5:
        for person in team:
            print(f"Welcome aboard {person}")
    else:
        print(f"Welcome All..!! {'-'.join(team)}")
gang_1 = ["Phoebe", "Monica", "Chandler", "Joey"]
gang_2 = ["Phoebe", "Monica", "Chandler", "Joey", "Rachel", "Ross"]

greet_team(gang_1)
greet_team(gang_2)

# if with in operator
# if with not in operator
print("Hi Ross" if "Ross" in gang_1 else "Ross is Missing")
print("Where's the Cake" if "Cake" not in "Birthday Party" else "That's really delicious")
```

## Rock-Paper-Scissors

```python
import random
game = ["rock", "paper", "scissors"]
try:
    while True:
        computer = game[random.randint(0,2)]
        user = input("Enter any one rock/paper/scissors: ").lower()
        if user not in game:
            raise
        elif computer == user:
            print(f"I have {computer}")
            print("Oh no!, That's a tie..")
        elif ((computer == "rock" and user == "scissors") or (computer == "paper" and user == "rock") or (computer == "scissors" and user == "paper")):
            print(f"I have {computer}")
            print("Yayy!, I won the game..")
        else:
            print(f"I have {computer}")
            print("Congrats!, You won the game..")
        cont = input("Do you want to continue playing? Say yes or no: ").lower()
        if cont == "no": break
except:
    print("Some error occurred!")
```

## Word Guessing Game

```python
print("I want you to guess a woman character name from Friends Television Series")
friends = ["PHOEBI", "MONICA", "RACHEL"]
guess_word = friends[random.randint(0,2)]
blank = ['_']*len(guess_word)
blanks = " ".join(blank)
for _ in range(0,6):
    i = input(f"The word is {blanks}, Enter a letter now: ").upper()
    if i in guess_word:    
        blank[guess_word.find(i)] = i
        blanks = " ".join(blank)
    else:
        print("That's not valid, Please try again..")
if "".join(blank)==guess_word:
    print("Well Done!, Nice Job..")
else:
    print("All your attempts are over now, Try again tomorrow!, Bye!!")
```
