# Task 1 - Inbuilt Data Structures and it's Methods

```python
print("**********List Methods**********")
l = [33,43,53,63,73,100,100]
print(l.append(83))
print(l.extend([1,2,3]))
print(l.copy())
print(l.count(100))
print(l.pop())
print(l.insert(1,20), l)
print(l.remove(73))
print(l.reverse(),l)
print(l.sort(),l)
print(l.sort(reverse=True),l)
print(l.clear())
print(len(l))

print("**********Tuple Methods**********")
t = "a",
print(t.count(1))
print(t.index("a"))
print(type(t))

print("**********Set Functions**********")
s1 = {1,2,5,4,10}
s2 = {1,3,4,6}
# print(s1.add(10),s1)
print(s1.difference(s2))
print(s1.intersection(s2))
print(s1.symmetric_difference(s2))

print("**********Dict Methods**********")
d1 = {"Apple":5, "Orange": 2, "Pomo": 6}
d2 = dict(one=1,two=2)
d3 = dict([(1,2),(3,4)])
print(d1,d2,d3)
d4 = {}
print(d4.fromkeys((100,99,98)))
print(d1.get("Pomo"))
for i,v in d1.items():
    print(i,v)
print(d1.keys(), d1.values(), d1.update([("Pomo",10)]), d1)
print(d1.pop("Orange"), d1.popitem(), d1.update([("Pomegranate",20)]), d1)






```
