## 1. Encapsulation

```python
class Employee:
    def __init__(self,name,salary):
        self.name = name
        self.__salary = salary # Private variable
    
    def get_salary(self):
        return self.__salary

# class TeamLead(Employee):
#     pass

emp1 = Employee("Alice", 1000)
# print(emp1.__salary) # Will throw error while accessing private attribute
print(emp1.get_salary()) # can be accessed via helper functions

class TeamLeader(Employee):
    pass

tl = TeamLeader("Mathew",2000)
# print(tl.__salary) # Error again
print(tl.get_salary())
        
```

## 2. Abstraction

```python
from abc import ABC, abstractmethod
class Employee(ABC):
    def __init__(self,name,salary):
        self.name = name
        self.__salary = salary # Private variable
    
    def get_salary(self):
        return self.__salary
    
    @abstractmethod
    def work_assigned(self):
        pass

# emp1 = Employee("Alice", 1000) # Error -> Cannot instantiate Abstract Class


class TeamLeader(Employee):
    def work_assigned(self):
        return "Here is you work"

tl = TeamLeader("Mathew",2000)
print(tl.work_assigned())
```

## 3. Inheritance

```python
class Employee:
    def __init__(self,name,salary):
        self.name = name
        self.__salary = salary # Private variable
    
    def get_salary(self):
        return self.__salary
    
    def work_assigned(self):
        return "From Parent Class"

class TeamLeader(Employee):
    def __init__(self, name, salary):
        super().__init__(name, salary)

    def team_members(self):
        return "Your team has 6 Devs and 2 QA"
    

tl = TeamLeader("Mathew",2000)
print(tl.work_assigned(),tl.team_members())

emp1 = Employee("Alice", 1000)
# print(emp1.team_members()) # Error Parent cannot access Child class methods
```

## 4. Polymorphism

```python
class Employee:
    def __init__(self,name,salary):
        self.name = name
        self.__salary = salary # Private variable
    
    def get_salary(self):
        return self.__salary
    
    def work_assigned(self):
        return "Generic Work - From Parent Class"

class TeamLeader(Employee):
    def __init__(self, name, salary):
        super().__init__(name, salary)

    def work_assigned(self): # Method Overriding -> mehtod from parent class is overridden by child
        return "Web Development - From Child Class"
    
    def rating(self,q1=3,q2=3,**kwargs): # Can handle many different number of inputs
        total_rating = 0
        for k,v in kwargs.items():
            total_rating += v
        total_rating = total_rating + q1 + q2
        return total_rating

emp1 = Employee("Alice", 1000)
print(emp1.work_assigned())
tl = TeamLeader("Mathew",2000)
print(tl.work_assigned())
print(tl.rating(5,3,certs=4,extra_contr=4,awards=3)) # multiple args
print(tl.rating()) # no-args

```
